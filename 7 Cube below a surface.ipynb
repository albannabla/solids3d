# HOW TO PLOT A SOLID MADE BY THE INTERSECTION OF A CUBE WITH A SURFACE

Let us start with a surface f(x,y):

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-1, 1, 10)
Y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = -np.sqrt(X**2 + Y**2)


fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z)
plt.show()
```

I will try to figure it out myself before browsing for solution on how to intersect a surface with a plane. 
Since I want this surface to be the top face of this cube, the problem will be mainly in solving one of the four side faces (and then iterate this solution four times), plus adding a bottom which is straight forward. 
I plan to use the polygon function that I have used for the previously plotted platonic solids.

Let us start from the face that is sitting on the yz plane. 
The curve goes from the vertex (-1,-1,-srqt(2)) to (1,-1,-srqt(2)).
I can sit the bottom face of the cube on the plane z = -2, so I also have the two bottom vertices: (-1,-1,-2) and (1,-1,-2).
Now I only have to find all of the points in the surface that have y = -1 and then I can plot this polygon. 

```python
print(Y)
```

The points seem to be the subset of coordinates given by (X[0],Y[0],Z[0]).
I can now put all of these coordinates in a variable. 

```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
front = []
for i in range(len(X[0])):
    front.append((X[0][i],Y[0][i],Z[0][i]))
front.append((1,-1,-2))
front.append((-1,-1,-2))
print(front)
```

Now let us try to plot this front face:

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

X = np.linspace(-1, 1, 10)
Y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = -np.sqrt(X**2 + Y**2)
front = []
for i in range(len(X[0])):
    front.append((X[0][i],Y[0][i],Z[0][i]))
front.append((1,-1,-2))
front.append((-1,-1,-2))

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-2, 1)
surf = ax.plot_surface(X, Y, Z)
c = Poly3DCollection([front]) 
ax.add_collection3d(c)
plt.show()
```

The back face is identical, just use Y = -1 and therefore (X[-1],Y[-1],Z[-1])

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

X = np.linspace(-1, 1, 10)
Y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = -np.sqrt(X**2 + Y**2)

front = []
for i in range(len(X[0])):
    front.append((X[0][i],Y[0][i],Z[0][i]))
front.append((1,-1,-2))
front.append((-1,-1,-2))

back = []
for i in range(len(X[-1])):
    back.append((X[-1][i],Y[-1][i],Z[-1][i]))
back.append((1,1,-2))
back.append((-1,1,-2))

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-2, 1)
c = Poly3DCollection([front,back]) 
c.set_facecolor(["yellow","red"])
ax.add_collection3d(c)
surf = ax.plot_surface(X, Y, Z)
plt.show()
```

Following the same concept, it is easy to do X = -1 and X = 1 as well.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

X = np.linspace(-1, 1, 10)
Y = np.linspace(-1, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = -np.sqrt(X**2 + Y**2)

front = []
for i in range(len(X[0])):
    front.append((X[0][i],Y[0][i],Z[0][i]))
front.append((1,-1,-2))
front.append((-1,-1,-2))

back = []
for i in range(len(X[-1])):
    back.append((X[-1][i],Y[-1][i],Z[-1][i]))
back.append((1,1,-2))
back.append((-1,1,-2))

left = []
for i in range(len(X)):
    left.append((X[i][0],Y[i][0],Z[i][0]))
left.append((-1,-1,-2))
left.append((-1,1,-2))

right = []
for i in range(len(X)):
    right.append((X[i][-1],Y[i][-1],Z[i][-1]))
right.append((1,-1,-2))
right.append((1,1,-2))

bottom = []
bottom.append((1,-1,-2))
bottom.append((1,1,-2))
bottom.append((-1,1,-2))
bottom.append((-1,-1,-2))

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-2, 1)
c = Poly3DCollection([front,back,left,right,bottom]) 
c.set_facecolor(["yellow","red","orange","green","brown"])
ax.add_collection3d(c)
surf = ax.plot_surface(X, Y, Z)
plt.show()
```
