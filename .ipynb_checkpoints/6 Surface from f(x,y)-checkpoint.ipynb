# PLOT A 3D SURFACE FROM A FUNCTION f(x,y)

The next step in my investigation of solids is to create an irregular solid made from functions.
I am thinking to firstly use a cube to define a rectangular region in 2D and intersect this with a function f(x,y) which exists at any point over that region. Before tackling that particular 3D solid I'd like to make sure that I am able to plot a surface in 3D starting from a function. 

I am thinking to use as a function z = ln(r) with r = sqrt(x^2 + y^2) for a region 1<x<2 and 1<y<2 (since ln(0) does not exist). 
Let us create a bunch of points and see if we can plot this: 

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(1, 2, 10)
Y = np.linspace(1, 2, 10)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.log(R)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z)
plt.show()
```

This worked out ok. Let us try a different function, trying to get rid of R and manipulate directly Z based on the meshgrid coordinates of X, Y.

As seen above, numpy allows us to directly work on the matrices without having to go through complicated looping, which saves considerable amount of time. 

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(0, 1, 10)
Y = np.linspace(0, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = np.exp(X+Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z)
plt.show()
```

Next I wanted to try to colour up these surfaces. I have found a brilliant page about the colormap function in mplot3d, here: 
http://mpastell.com/2013/05/02/matplotlib_colormaps/
I will try to use one of these:

```python
from pylab import colormaps
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(0, 1, 10)
Y = np.linspace(0, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = np.exp(X+Y)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=colormaps()[2])
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

This is quite nice and it can be seen how the colour gets darker for higher value of z.
Please note that colormap() is not a function but a LIST, so it can be used by indexing the correct colorscheme desired. The full list is on the link above.

Another way to use colormaps is to use the library cm:
https://matplotlib.org/3.2.1/gallery/color/colormap_reference.html
Some of the two libraries are identical color scheme and each of these can be reversed by adding _r to the name 

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(0, 1, 10)
Y = np.linspace(0, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = np.exp(X+Y)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.copper_r)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

Now, it would be interesting to "add a dimension" to the plot and change the colours based on a different function with 3 independent variables. 
I think this can be done using pcolormesh attribute of ax.
https://matplotlib.org/gallery/images_contours_and_fields/pcolormesh_levels.html#sphx-glr-gallery-images-contours-and-fields-pcolormesh-levels-py

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(0, 1, 10)
Y = np.linspace(0, 1, 10)
X, Y = np.meshgrid(X, Y)
Z = np.exp(X+Y)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
turf = ax.pcolormesh(X, Y, Z, cmap=cm.coolwarm)
fig.colorbar(turf, shrink=0.5, aspect=5)
plt.show()
```
